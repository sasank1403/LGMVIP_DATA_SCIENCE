# **IMAGE TO PENCIL SKETCH**

```python
import cv2
from google.colab.patches import cv2_imshow 
```

**UPLOADING THE IMAGE**

```python
image=cv2.imread("/content/Labrador_Retriever_(1210559).jpg")
original_image =cv2.resize(image, (700, 500))
cv2_imshow(original_image)
```

**CONVERT RGB SCALE IMAGE INTO GRAY SCALE IMAGE**

```python
gray_scale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_image)
```

**INVERTED THE GRAY SCALE IMAGE**

```python
inverted_gray_scale_image = cv2.bitwise_not(gray_scale_image)
cv2_imshow(inverted_gray_scale_image)
```

**APPLY GUASSIAN BLUR**

```python
inverted_blurry_image = cv2.GaussianBlur(inverted_gray_scale_image, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(inverted_blurry_image)
     
```

**MIXING THE GRAY SCALE IMAGE WITH INVERTED BLURRY IMAGE**

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_pencil_sketch_image = dodgeV2(gray_scale_image,inverted_blurry_image)

cv2_imshow(final_pencil_sketch_image )
```

**ORIGINAL IMAGE VS PENCIL SKETCH IMAGE**

```python
cv2_imshow(original_image)

cv2_imshow(final_pencil_sketch_image)
```
